```python
file = 'Test2-data.xlsx'
```

```python
df = pd.read_excel(file, skiprows=1)
```

```python
df
```

```python
Atlantic_Canada = ["Newfoundland and Labrador", "Prince Edward Island", "Nova Scotia",
"New Brunswick"]
```

```python
df = df[df['Province/Territory'].isin(Atlantic_Canada)]
df
```

```python
# Determine the Top 10 most popular requested occupations from the Occupation field by count
```

```python
df['Occupation']
```

```python
count = df['Occupation'].value_counts()
count
```

```python
top_10 = count.head(10)
top_10
```

```python
# Determine the sum of the rest of the occupations applied for and refer to it as XXXX-Other.
```

```python
XXXX_Other = count[11:].sum()
XXXX_Other
```

```python
pd.Series({'XXXX-other': XXXX_Other})
```

```python
top_10_and_other = pd.concat([top_10, pd.Series({'XXXX-other': XXXX_Other})])
```

```python
top_10_and_other
```

```python
plt.figure(figsize=(6,5))
plt.pie(top_10_and_other, labels = top_10_and_other.index)
plt.title('Top 10 Occupations in Atlantic Canada, Q3-2023')
```
